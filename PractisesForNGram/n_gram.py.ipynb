```python
def tranin_char_lm(data , n):
    #data = open(filename).read()
    lm = defaultdict(Counter)
    pad = "~" * n
    data = pad + data
    for i in range(len(data) - n):
        history , char = data[i : i + n] , data[i + n]
        #print("History : " , history , sep = "\n")
        lm[history][char] += 1
        #print(lm[history][char])
    def normalize(counter):
        #print(counter)
        s = float(sum(counter.values()))
        #print(s)
        return [(c , cnt/s) for c , cnt in counter.items()]
    outlm = {hist:normalize(chars) for hist , chars in lm.items()}
    return outlm
```

```python
def generate_letter(lm , history , order):
    history = history[-order:]
    dist = lm[history]
    x = random()
    for c, v in dist:
        x -= v
        if x <= 0: 
            return c
```

```python
def generate_text(lm , order , letters= 5000):
    history = "~" * order
    out = list()
    for i in range(letters):
        c = generate_letter(lm , history , order)
        history = history[-order:] + c
        out.append(c)
    return out
```

```python
import pickle
def save_obj(obj, name ):
    with open(name + '.pkl', 'wb') as f:
        pickle.dump(obj, f, pickle.HIGHEST_PROTOCOL)

def load_obj(name ):
    with open(name + '.pkl', 'rb') as f:
        return pickle.load(f)
```

```python
def generate_ngrams(words_list, n):
    ngrams_list = []
 
    for num in range(0, len(words_list)):
        ngram = ' '.join(words_list[num:num + n])
        ngrams_list.append(ngram)
 
    return ngrams_list
```

```python
def generate_char_ngram(data, n):
    ngrams_list = []
    for i in range(len(data)):
        ngram = ''.join(data[i:i + n])
        ngrams_list.append(ngram)
    return ngrams_list
```

```python
def process_text(text):
 
    #text = text.lower()
    text = text.replace(',', ' ')
    text = text.replace('/', ' ')
    text = text.replace('(', ' ')
    text = text.replace(')', ' ')
    text = text.replace('.', ' ')
    text = text.replace('\'' , '')
 
    # Convert text string to a list of words
    return text#.split() split for word
```

```python
def create_word_all_grams_with_probabilies(data, till=3 , function = generate_ngrams):
    all_grams = list()
    sorted_grams = list()
    data = process_text(data) #for word n gram
    for i in range(till + 1):
        grams = function(data, i)
        counts = Counter(grams)
        probs = {gram : (counts[gram]/len(counts)) for gram in counts.keys()}
        sorted_grams.append(sorted(probs.items(), key=lambda kv: kv[1] , reverse=True))
        all_grams.append(probs)
    return all_grams , sorted_grams
```

```python
def calculate_perplexity_for_given_text(lm, text , n):
    #tokens = text.split() #for word
    tokens = text
    result = 1.0
    if (len(tokens) > 3):
        power = 3
    else:
        power = len(tokens)
    power = n # for char
    for i in range(power):
        result *= lm[i + 1].get(''.join(tokens[:i + 1]))
    result = (1 / result) ** (1 / power)
    return result
```
