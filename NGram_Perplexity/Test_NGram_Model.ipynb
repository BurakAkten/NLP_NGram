```python
char_model_1 = NGram.load_model(char_model_1_name)
char_model_2 = NGram.load_model(char_model_2_name)
char_model_3 = NGram.load_model(char_model_3_name)
char_model_4 = NGram.load_model(char_model_4_name)
char_model_5 = NGram.load_model(char_model_5_name)
```

```python
syll_model_1 = NGram.load_model(syll_model_1_name)
syll_model_2 = NGram.load_model(syll_model_2_name)
syll_model_3 = NGram.load_model(syll_model_3_name)
syll_model_4 = NGram.load_model(syll_model_4_name)
syll_model_5 = NGram.load_model(syll_model_5_name)
```

```python
test_text = "anayasa"
```

```python
def print_Perplexty(models , test_text):
    models = models[::-1]
    for i , model in enumerate(models):
        print("{}-Gram".format(i + 1) , model.perplexity_of(test_text).split(":")[1])
```

```python
def print_probability(models , test_text):
    models = models[::-1]
    for i,model in enumerate(models):
        print("{}-Gram".format(i + 1) , model.probability_of(test_text).split(":")[1])
```

```python
print_Perplexty([syll_model_1,syll_model_2,syll_model_3,syll_model_4,syll_model_5] , test_text)
```

```python
print_Perplexty([char_model_1,char_model_2,char_model_3,char_model_4,char_model_5] , test_text)
```

```python
print_probability([syll_model_1,syll_model_2,syll_model_3,syll_model_4,syll_model_5] , test_text)
```

```python
print_probability([char_model_1,char_model_2,char_model_3,char_model_4,char_model_5] , test_text)
```
